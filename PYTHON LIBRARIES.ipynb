```python
import pandas as pd
df=pd.read_excel("np.xlsx")
mind=df["price"].min()
maxd=df["price"].max()
df["Noemalization"]=(df["price"] - mind) / (maxd - mind)
print(df[["price","Noemalization"]])
```

```python
import pandas as pd
df=pd.read_excel("np.xlsx")
std=df["price"].std()
mean=df["price"].mean()
df["stdar"]=(df["price"] - mean)/std
print(df[["price","stdar"]])
```
